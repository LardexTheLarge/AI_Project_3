# Emission Data EDA

```python
#import libraries
import pandas as pd
import seaborn as sns
import project_libs as libs
import matplotlib.pyplot as plt
import warnings 
```

```python
#Load master emission data
master_data = libs.read_csv_to_dataframe('data/master_emission_data.csv')
```

```python
#Format year as date and time
master_data['Year'] = pd.to_datetime(master_data['Year'], format='%Y')
```

```python
#view missing values
master_data.head()
```

```python
fig, axs = plt.subplots(3, 2, figsize=(16, 12))
# Plot 1 - Annual greenhouse gas emissions in CO₂ equivalents
sns.lineplot(data=master_data, x='Year', y='Annual greenhouse gas emissions in CO₂ equivalents', ax=axs[0, 0])
axs[0, 0].set_title('Annual GHG Emissions')

# Plot 2 - Cap_and_trade
sns.lineplot(data=master_data, x='Year', y='temp_change', ax=axs[0, 1])
axs[0, 1].set_title('Temprature change')

# Plot 3 - Emission_target
sns.lineplot(data=master_data, x='Year', y='emission_target', ax=axs[1, 0])
axs[1, 0].set_title('Emission Target')

# Plot 4 - Renewable_energy
sns.lineplot(data=master_data, x='Year', y='renewable_energy', ax=axs[1, 1])
axs[1, 1].set_title('Renewable Energy')

# Plot 5 - emission vs tem change
sns.scatterplot(data=master_data, x='Annual greenhouse gas emissions in CO₂ equivalents', y='temp_change', ax=axs[2, 0])
axs[2, 0].set_title('Temp Change vs Emission')

# Plot 6 - emission vs green house energy
sns.scatterplot(data=master_data, y='Annual greenhouse gas emissions in CO₂ equivalents', x='renewable_energy', ax=axs[2, 1])
axs[2, 1].set_title('Renewable energy vs Emission')

plt.suptitle('Greenhouse Gas Analysis', fontsize=26)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
