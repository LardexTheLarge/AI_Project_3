### Forecast Emission, emission target, green energy and predict temperature change

```python
#import required libraries.
import pandas as pd
import project_libs as libs
from prophet import Prophet
from datetime import datetime
```

```python
#Load Emission Data
master_data = libs.read_csv_to_dataframe('data/master_emission_data.csv')
master_data = master_data.rename(columns={'Annual greenhouse gas emissions in CO₂ equivalents':'emission'})
```

```python
#Change Year from int to datetime format
master_data['Year'] = master_data['Year'].apply(lambda year: datetime(int(year), 12, 31))

#extract all numrical data with year
yearly_emission,yearly_cap,yearly_target,yearly_temp_chamge, yearly_clean_energy ,yearly_tem_change= master_data[['Country','Year','emission']],master_data[['Country','Year','Cap_and_trade']],\
    master_data[['Country','Year','emission_target']],master_data[['Country','Year','temp_change']],master_data[['Country','Year','renewable_energy']], master_data[['Country','Year','temp_change']]

```

```python
#View head
master_data.head()
```

```python
#Change column names to ds and y for prophet
f_yearly_emission = yearly_emission.rename(columns={'Year':'ds','emission':'y'})
f_yearly_cap = yearly_cap.rename(columns={'Year':'ds','Cap_and_trade':'y'})
f_yearly_target = yearly_target.rename(columns={'Year':'ds','emission_target':'y'})
f_yearly_clean_energy= yearly_clean_energy.rename(columns={'Year':'ds','renewable_energy':'y'})
```

```python
df_list  = [f_yearly_emission,f_yearly_cap,f_yearly_target,f_yearly_clean_energy]
df_name  = ['yearly_emission','yearly_cap','yearly_target','yearly_clean_energy']
df_col   = [yearly_emission,yearly_cap,yearly_target,yearly_clean_energy]
forecast_df_list = []

# Loop through each DataFrame
for idx, df in enumerate(df_list):
    print(f"Forecasting for DataFrame {df_name[idx]}")
    print('+'* 100)
    # Initialize Prophet model
    col_list = ['Country','Year',df_col[idx].columns[2]]
    col_names = {'ds':'Year', 'yhat':df_col[idx].columns[2]}
    final_data = pd.DataFrame(columns= col_list)
    country_list = df_list[idx]['Country'].unique().tolist()
    country_df = pd.DataFrame(columns= ['Country','Year',df_col[idx].columns[2]])
    print(country_list)
    
    for country in country_list:
        df = df_list[idx][df_list[idx]['Country'] == country][['ds','y']]
        model = Prophet()

    # Fit the model
        model.fit(df)

    # Create future dates for forecasting
        future = model.make_future_dataframe(periods = 10, freq = 'YE')  # Forecast for 365 days  
    # Make predictions
        print(f"predictiing {df_name[idx]} for {country}")
        forecast = model.predict(future)
        forecast = forecast.rename(columns=col_names)
        print('-'* 1000)
        forecast['Country'] = country
        forecast = forecast[col_list]
        #final_data = final_data.append(forecast, ignore_index = True)
        country_df = pd.concat([country_df, forecast])
    # Assign the forecast results to a new DataFrame with a dynamic name
    #print(country_df)
    forecast_df_name = f"forecast_{df_name[idx]}"  # Create a dynamic name for the forecast DataFrame
    
    globals()[forecast_df_name] = country_df# Extract relevant columns (ds for dates, yhat for predictions)
    forecast_df_list.append(globals()[forecast_df_name])
    

    
```

```python
#Combine all forecasted data into one data frame
Forecasted_data = pd.merge(forecast_df_list[0], forecast_df_list[1], on=['Year','Country'], how='left')
Forecasted_data = pd.merge(Forecasted_data, forecast_df_list[2], on=['Year','Country'], how='left')
Forecasted_data = pd.merge(Forecasted_data, forecast_df_list[3], on=['Year','Country'], how='left')
Forecasted_data = pd.merge(Forecasted_data, yearly_tem_change, on=['Year','Country'], how='left')
```

```python
#Save forecasted data 
Forecasted_data.to_csv('data/forecasted_master_data.csv', index = False)
```
